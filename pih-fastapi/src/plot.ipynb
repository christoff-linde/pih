```python
import pandas as pd
df = pd.read_csv("data/data.csv")
df.head()
```

```python
import datetime


# add column to dataframe with formatted timestamps
df["datetime"] = df.apply(
    lambda row: datetime.datetime.fromtimestamp(row["timestamp"]).strftime(
        "%Y-%m-%d %H:%M:%S"
    ),
    axis=1,
)
df.head()
```

```python
import plotly.express as px

fig = px.line(df, x="datetime", y="temperature")
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=df["datetime"], y=df["temperature"], name="temperature"))
fig.add_trace(go.Scatter(x=df["datetime"], y=df["humidity"], name="humidity"))
fig.show()
```
